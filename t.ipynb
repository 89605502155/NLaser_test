```python
Xdata = pkl.load(open('X_new.pkl.gz', 'rb'))
Ydata = pkl.load(open('y.pkl.gz', 'rb'))
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
model=NLaser(
    n_components=4, a=0.09
)
y=Ydata['Gominy'].to_numpy()
model.fit(xx=Xdata['X'],yy=y)
def flatten(x):
    sh = x.shape
    return x.reshape(sh[0:len(sh)-1])
print(flatten(model.w_i).transpose()[:,0])
print(flatten(model.w_k).transpose()[:,0])
plt.figure()
plt.plot(flatten(model.w_i).transpose())

plt.figure()
plt.plot(flatten(model.w_k).transpose())

```
